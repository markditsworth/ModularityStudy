```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m1=3
m2=0.6
m=1.7
```

```python
x1 = np.random.random(size=50)*0.4
y1 = m1*x1 + (np.random.random(size=50)-0.5)*.5 + 0.1
x2 = np.random.random(size=50)
y2 = m2*x2 + (np.random.random(size=50)-0.5)*.2 + 0.1
```

```python
x11 = np.arange(0,0.8,0.2)
x22 = np.arange(0.24,1,0.2)
d1 = x11*m+0.4
d2 = x22*m-0.4
x = np.arange(0,1,0.2)
d = x*m
```

```python
x1_in = x1[y1 < x1*m+0.4]
y1_in = y1[y1 < x1*m+0.4]
x1_out= x1[y1 >= x1*m+0.4]
y1_out= y1[y1 >= x1*m+0.4]

x2_in = x2[y2 >= x2*m-0.4]
y2_in = y2[y2 >= x2*m-0.4]
x2_out= x2[y2 < x2*m-0.4]
y2_out= y2[y2 < x2*m-0.4]
```

```python
plt.plot([0,0],[0,1.4],c='k')
plt.plot([0,1],[0,0],c='k')
plt.scatter(x1_in,y1_in,color='k',s=10)
plt.scatter(x1_out,y1_out,marker='x',color='k',s=30)
plt.scatter(x2_in,y2_in,color='k',s=10)
plt.scatter(x2_out,y2_out,marker='x',color='k',s=30)
plt.plot(x11,d1,color='k',ls='--')
plt.plot(x22,d2,color='k',ls='--')
plt.plot(x,d,ls='-',color='k')

A=0.55
B=0.8
xx=np.linspace(0.36,0.6,1000)
yy = np.sqrt(1 - (xx/A)**2)*B
plt.plot(xx,yy,c='k')
plt.text(0.54,0.18,r'$\theta$',fontsize=14)

plt.text(0.8,1.4,r'$\mathbb{L}$',fontsize=16)
plt.gca().annotate('',(0.58,1),(0.5,1.23),ha="center",va="center",arrowprops=dict(arrowstyle='|-|'))
plt.text(0.55,1.15,r'$d$',fontsize=14)
```
